```python
# Portfolio Optimaization  Equal weight performance
Stocks = pd.concat([Divis['Adj Close'],BergerPaints['Adj Close'],Astral['Adj Close'],AsianPaints['Adj Close'],ABcapital['Adj Close'],Idfcfirst['Adj Close'],Jubilant['Adj Close'],IGL['Adj Close'],Pidilite['Adj Close'],Lalpathlab['Adj Close'],HdfcBank['Adj Close'],Hdfclife['Adj Close'],Alkyl['Adj Close'],Titan['Adj Close'],Vinati['Adj Close'],Galaxy['Adj Close'],Motherson['Adj Close'],Relaxo['Adj Close'],HCL['Adj Close'],Kotak['Adj Close']],axis=1)
Stocks.columns = ['Divis','BergerPaints','Astral','AsianPaints','ABcapital','Idfcfirst','Jubilant','IGL','Pidilite','Lalpathlab','HdfcBank','Hdfclife','Alkyl','Titan','Vinati','Galaxy','Motherson','Relaxo','HCL','Kotak']
Stocks.head()
```

```python
# Converting the return  Equal weight performance
Stock_return = Stocks.pct_change(1)
Stock_return.head()
```

```python
#Setting up 10k different portfolio weighting combination  Equal weight performance
number_ports = 10000
all_weight =  np.zeros((number_ports, len(Stocks.columns)))
all_weight
#len(Stocks.columns)
```

```python
# Allocating space for portfolio calculations - to determine portfolio returns of 10,000 iteration for all the scenarios  
returns_array = np.zeros(number_ports)
volatility_array = np.zeros(number_ports)
sharpe_array = np.zeros(number_ports)
```

```python
# for replication from randomsizer  Equal weight performance
import random
random.seed(3)
# ramdomsizer for array 
np.random.seed(3)
# for loop for different combination 
for index in range(number_ports): 
    #generate the random weights 
     number = np.array(np.random.random(20)) 
     weights = number/np.sum(number)
    
    #save_weights
     all_weight[index, :] = weights
    #expected returns
     returns_array[index] = np.sum(Stock_return.mean() * 252 * weights)
    #Expected Volatility
     volatility_array[index] = np.sqrt(np.dot(weights.T, np.dot(Stock_return.cov() * 252,weights)))
    #sharpe ratio
     sharpe_array[index] = returns_array[index]/volatility_array[index]
        
        
    
```

```python
# Print all the weight combination  Equal weight performance
print("All weights:",all_weight)
#print first weight
print("First combination:",all_weight[0])
```

```python
#Print all sharpe ratio  Equal weight performance
print("All the sharpe ratio:",sharpe_array)

#print first sharpe
print("Sharpe ratio for first index:",sharpe_array[0])
```

```python
#Find the portfolio with highest sharpe ratio  Equal weight performance
sharpe_array.max()
```

```python
#Find by the index
index_max_sharpe = sharpe_array.argmax()
index_max_sharpe
```

```python
#Taiking the best weights of portfolio
print(all_weight[index_max_sharpe,:])

print(Stocks.columns)
```
